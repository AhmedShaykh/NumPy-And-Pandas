```python
def fnObtained(s1: int, s2: int, s3: int, s4: int, s5: int) -> int:

    return s1 + s2 + s3 + s4 + s5;

df1["Obtained"] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x: fnObtained(* x), axis=1); # Single Output & Multi Input

df1
```

```python
def fnObtained(s1: int, s2: int, s3: int, s4: int, s5: int) -> tuple[int, int, float, str]:

    total: int = 500;

    obtained :int = s1 + s2 + s3 + s4 + s5;

    per : float = obtained / total * 100;

    grade : str = "";

    if per >= 80:
        grade = "A+";
    elif per >= 70:
        grade = "A";
    elif per >= 60:
        grade = "B";
    elif per >= 50:
        grade = "C";
    elif per >= 40:
        grade = "D";
    elif per >= 33:
        grade = "E";
    else:
        grade = "F";

    return total, obtained, per, grade;

df1[["Total","Obtained", "Percentage", "Grade"]] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x: fnObtained(* x),
                                                                                             axis=1,
                                                                                             result_type="expand");

display(df1); # Multi Output & Multi Input
```

```python
from typing import Dict;

remark : Dict[str, str] = {
    "A+" : "Excellent",
    "A" : "Very Good",
    "B" : "Good",
    "C" : "Fair",
    "D" : "Poor",
    "E" : "Very Poor",
    "F": "Fail"
};

df1["Remark"] = df1.Grade.map(remark);

display(df1);
```
