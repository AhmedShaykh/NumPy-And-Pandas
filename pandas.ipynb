# **Slicing & Indexing**

```python
s1 : pd.Series = pd.Series([1,2,3,4,5]);

display(s1);

display(s1[1]);
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5]);

display(s1);

print("Applying Slicing");

display(s1.iloc[1:4]); # Index Location Pass Only Numbers Same As Numpy Slicing
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5], index=["A", "B", "C", "D","E"]);

display(s1);

print("Applying Slicing");

display(s1.iloc[1:4]);
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5], index=["A", "B", "C", "D","E"]);

display(s1);

print("Applying Slicing");

display(s1.loc["A":"D"]); # Index Location Pass Labels & Included End Value
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5], index=["A", "B", "C", "D","E"]);

display(s1);

print("Applying Slicing");

display(s1.iat[1]); # Index Location Pass Number Extract Only One Cell Value & Update It
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5], index=["A", "B", "C", "D","E"]);

display(s1);

print("Applying Slicing");

display(s1.at["D"]); # Index Location Pass Label Extract Only One Cell Value & Update It
```
