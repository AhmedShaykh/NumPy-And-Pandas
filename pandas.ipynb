# **Series**

```python
s1 : pd.Series = pd.Series([1,2,3,4,5]); # Pandas Series

print(s1);
```

```python
s1 : pd.Series = pd.Series((1,2,3,4,5));

print(s1);
```

```python
s1 : pd.Series = pd.Series({
                            "a": 10,
                            "b":20,
                            "c":30,
                            "d":40,
                            "e":50,
                            "f":60,
                            "g":70,
                        }); # Series Apply In Tuple, Dict & List But Not Apply In Set

print(s1);
```

```python
values: list[int] = [1, 2, 3, 4, 5];

indexs: list[str] = ["a", "b", "c", "d", "e"];

s1: pd.Series = pd.Series(values, index=indexs);

print(s1);
```

```python
values: list[int] = [1, 2, 3, 4, 5];

indexs: list[list[str]] = [
                        ["a1", "a1", "a1", "b1", "b1"], # Multi Index
                        ["a", "b", "c", "d", "e"]
                    ];

s1: pd.Series = pd.Series(values, index=indexs);

print(s1);
```

```python
values: list[int] = [1, 2, 3, 4, 5];

indexs: list[list[str]] = [
                        ["a1", "a1", "a1", "b1", "b1"],
                        ["a", "b", "c", "d", "e"]
                    ];

s1: pd.Series = pd.Series(values, index=indexs , name="Pandas Series");

print(s1);
```

```python
import numpy as np;

values: list[int] = [1, 2, 3, 4, 5];

indexs: list[list[str]] = [
                        ["a1", "a1", "a1", "b1", "b1"],
                        ["a", "b", "c", "d", "e"]
                    ];

s1: pd.Series = pd.Series(values,
                          index=indexs,
                          name="Pandas Series",
                          dtype=np.int32);

print(s1);
```

# **DataFrame**

```python
s1 : pd.Series = pd.Series([1,2,3,4,5], name="ID");

s2 : pd.Series = pd.Series([10,20,30,40,50], name="Score");

s3 : pd.Series = pd.Series(["Suleman","Ahsan","Junaid","Ahmed","Abdullah"], name="Developer Name");

df : pd.DataFrame = pd.DataFrame({"ID": s1, "Score": s2, "Developer Name": s3}); # Pandas DataFrame

df
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5]);

s2 : pd.Series = pd.Series([10,20,30,40,50]);

s3 : pd.Series = pd.Series(["Suleman","Ahsan","Junaid","Ahmed","Abdullah"]);

df : pd.DataFrame = pd.DataFrame({"ID": s1, "Score": s2, "Developer Name": s3});

df
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5], name="ID");

s2 : pd.Series = pd.Series([10,20,30,40,50], name="Score");

s3 : pd.Series = pd.Series(["Suleman","Ahsan","Junaid","Ahmed","Abdullah"], name="Developer Name");

df : pd.DataFrame = pd.DataFrame([s1, s2, s3]);

df
```

```python
s1 : pd.Series = pd.Series([1,2,3,4,5], name="ID");

s2 : pd.Series = pd.Series([10,20,30,40,50], name="Score");

s3 : pd.Series = pd.Series(["Suleman","Ahsan","Junaid","Ahmed","Abdullah"], name="Developer Name");

df : pd.DataFrame = pd.concat([s1, s2, s3], axis=1);

df
```
